The idea is to do a generalized network creation routine that uses geojson

```python
import pysal as ps
N = ps.Network
```

```python
import inspect
```

```python
print(inspect.getsource(N._extractnetwork))
```

```python
def _extractnetwork(self):
    
    C = ps.cg.shapes.Chain
    
    #C.__repr__ =  chainrepr
    
    self.chains = {}
    
    nodecount = 0
    shps = ps.open(self.in_shp)
    for shp in shps:
        vertices = shp.vertices
        start = vertices[0]
        stop = vertices[-1]

        #Get a unique ID for the start and stop nodes if new, else get the original id
        try:
            vid = self.nodes[start]
        except:
            self.nodes[start] = vid = nodecount
            nodecount += 1
        
        try:
            nvid = self.nodes[stop]
        except:
            self.nodes[stop] = nvid = nodecount
            nodecount += 1
        
        self.adjacencylist[vid].add(nvid)
        self.adjacencylist[nvid].add(vid)
        
        edgenodes = sorted([vid, nvid])
        edge = tuple(edgenodes)
        self.edges.append(edge)
        self.chains[edge] = C(vertices)
        self.edge_lengths[edge] = self.chains[edge].len
    print nodecount
```

```python
#Patch in the new _extractnetwork
N._extractnetwork = _extractnetwork

#Sanity Check
print(inspect.getsource(N._extractnetwork))
```
